  # Question Three: Missing Number in Range
 
  You are given a list containing `n` integers in the range `[0, n]`. Return a list of numbers that are missing from the range `[0, n]` of the array. If there is no missing number, return -1. Note, all the integers in the list may not be unique.
  
  ## Examples

  ### Example 1

  Input: `lst = [0, 2]`

  Output: [1]

  ### Example 2

  Input: `lst = [5, 0, 1]`

  Output: [2, 3, 4]

  ### Example 3

  Input: `lst = [6, 8, 2, 3, 5, 7, 0, 1, 10]`

  Output: [4, 9]

</details>

#### Starter Code for Question 3


```python
#Since starter code uses List, need to import module
from typing import List

def missing_num(nums: List) -> int:
    #Create case for empty list, in case. Returns [-1].
    if not nums:
        return [-1]
    
    #Get maximum value of list to create loop
    max_value = max(nums)

    #Populate full range of [0, n] for comparison, where n is max_value. Start with an empty list 'full_range' and append to it based on counter 'i'.
    full_range = []
    i = 0

    while i <= max_value:
        full_range.append(i)
        i += 1

    #Loop through every value in full_range, and check if it is in input list 'nums'. If not, we add it to missing_numbers list.
    missing_numbers = []
    for int in full_range:
        if int not in nums:
            missing_numbers.append(int)
    
    #Build condition where missing_numbers is empty, with non-empty input 'nums' List. Returns [-1].
    if len(missing_numbers) == 0:
        return [-1]

    #Return missing_numbers list as output
    return missing_numbers


missing_num([0, 1, 2, 3, 9, 7, 4])
```

```python
# Your answer here
#Input: s = "([{}])"
#Output: True
#This returns true because every bracket is accounted for, and paired in the correct order.
#each pair of outer bracket is paired with the correct closing one in order to match fully.


#One example Ghazaleh made is:
#  Input: s = "((())"
#  Output: False
# This makes sense because there is an extra opening bracket, which will have nothing to properly pair with.
```


-   Copy the solution your partner wrote. 


```python
# Your answer here

#Code directly copied from partner below:
def is_valid_brackets (s:str) -> bool:
    stack = []
    bracket_map = {')': '(', '}': '{', ']': '['}

    for char in s:
        if char in bracket_map.values():
            stack.append(char)
        elif char in bracket_map:
            if not stack or stack[-1] != bracket_map[char]:
                 return False
            stack.pop()
        else:
            return False
    return not stack
```


-   Explain why their solution works in your own words.


```python
# Your answer here

#Ghazaleh's solution essentially creates a queue of bracket pairs, and does a "match-em" for each one:
#   - It starts by defining a dictionary for each pair of brackets, and creating an empty stack to perform the pair matching 
#   - It loops through every character in the input string:
#       - If a character is an opening bracket, it adds it to the stack.
#       - If a character is a closing bracket, it checks that the last opening bracket matches the current closing bracket character.  
#           - it also checks that the current stack isn't empty (i.e. it can't be an empty stack and then add a closing bracket)
#           - if either of these conditions fail, return False - the string is not a valid bracket string
#           - if the closing bracket matches the opening bracket in the stack, it gets popped off (like a "match-em" game)
#       - The function also has a checking case for any invalid characters (non-brackets). If they aren't a valid bracket character from the dictionary, it returns False.

#The solution works because:
#   - ensures that all bracket characters are accounted for
#   - it ensures the brackets are being closed in the correct order and in the correct pairing.
#   - also checks for invalid characters (non-bracket characters)
```
