
-   Code the solution to your assigned problem in Python (code chunk). Note: each problem can be solved more simply if you use an abstract data type that is suitable for that problem. Using that try to find the best time and space complexity solution!


````python
# Your answer here

#Copied from above, removed 'pass', and completed below.
from typing import List

def move_zeros_to_end(nums: List[int]) -> List[int]:
    
    #Create empty return list to fill out non-zero values in order, and another to hold all zero values.
    end_zeros = []
    non_zeros = []   

    # loop through each number in the input list. 
    for num in nums:
        #if the number is 0, add it to the end_zeros list.
        if num == 0:
            end_zeros.append(num)
        #otherwise add it to the non_zeros list.
        else:
            non_zeros.append(num)
    
    #concatenate the lists together
    output = non_zeros + end_zeros
    return output

#Test the function
move_zeros_to_end([0, 1, 0, 3, 12])
````

````python
# Your answer here

#The time conmplexity is O(n):
#   - The for loop for the input list is O(n), since it goes through all n elements once.
#   - The condition to check if the element is 0 (if num == 0) and the append functions are O(1)
#   - THe concatenation at the end AT WORST will be O(n) - for example if all elements are zero or nonzero, so it is O(n).
#       The worst case scenario will only occur for one of the lists (end_zeros or non_zeros) at any given time.
# THe result is O(n) + O(n), which is just O(n) for time complexity.

#The space complexity is also O(n)
#   - At most, each of the intermediate lists (non_zeros, end_zeros) will only be size n. This occurs when all elements in input list are zero or non-zero.
#   - The output list will always be size n.  
````
